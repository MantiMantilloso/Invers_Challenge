# NYC Taxi Trips Exploration

```python
# Import libraries to read
import pandas as pd

# try to read a single file (green taxis):
trips = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2022-01.parquet')
trips.head()
```

```python
# Diference with green from other year?
trips2 = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-01.parquet')
trips2.head()
```

```python
print(f'2022 has: {trips.shape}')
print(f'2025 has: {trips2.shape}')
```

## Diferencias detectadas en 2 ejemplos:
- ehail_fee in 2022 was null vs 2025 was NaN
- 2022 has 20 cols, 2025 has 21 cols (+ cbd_congestion_fee)


```python
common_cols = list(set(trips.columns) & set(trips2.columns))
diff_cols_trips = list(set(trips.columns) - set(trips2.columns))
diff_cols_trips2 = list(set(trips2.columns) - set(trips.columns))

print(f"Columns unique to 'trips': {diff_cols_trips}")
print(f"Columns unique to 'trips2': {diff_cols_trips2}")
print(f"Common columns: {common_cols}")
```

## Now let's check for the yellow trip data

```python
# try to read a single file (yellow taxis):
ytrips = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2022-01.parquet')
ytrips.head()
```

```python
ytrips2 = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2025-01.parquet')
ytrips2.head()
```

```python
print(f'yellow 2022 has: {ytrips.shape}')
print(f'yellow 2025 has: {ytrips2.shape}')
```

```python
ycommon_cols = list(set(ytrips.columns) & set(ytrips2.columns))
ydiff_cols_trips = list(set(ytrips.columns) - set(ytrips2.columns))
ydiff_cols_trips2 = list(set(ytrips2.columns) - set(ytrips.columns))

print(f"Columns unique to 'trips': {ydiff_cols_trips}")
print(f"Columns unique to 'trips2': {ydiff_cols_trips2}")
print(f"Common columns: {ycommon_cols}")
```
